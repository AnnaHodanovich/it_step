#     Задание: придумать и создать два класса с наследованием у родителя есть два атрибута и метода, а у дочернего добавить ещё один метод
   

Создаем два класса:
Класс-родитель: "Посетитель магазина" с атрибутами - пол, размер, рост, предпочтения по бренду и цвету, количество денег

Дочерний класс: "Покупатель в магазине", который будет унаследовать атрибуты класс-родителя - пол, размер, рост, а также иметь свои дополнительные параметры - деньги (наличные или безналичные).

Методы класса-родителя: примерять одежду, выбирать цвет одежды

Методы дочернего класса: +оформлять покупку/расчитывать деньгами

Можно ввести дополнительный класс - "пальто", которое и будут примерять и покупать наши два класса

```python
class Coat(object):
        #аттрибуты экземпляра класса-родителя
    def __init__(self, sex: str = 'male', size: str = 'S', brand: str = 'Mango', color: str = 'blue', price: int=0):
        self.size = size
        self.sex = sex
        self.brand = brand
        self.color = color
        self.price = price
```

```python
class Vizitor(object):
    #аттрибуты экземпляра класса-родителя
    def __init__(self, sex: str = 'male', size: str = 'S', brand: str = 'Mango', color: str='blue'):
        self.size = size
        self.sex = sex
        self.brand = brand
        self.color = color
        
    #методы порождаемого экземпляра класса
    def put_on_coat(self, coat):
        
        if self.size==coat.size and self.sex==coat.sex:
            print(f"{coat.size} is my size. It suits good")
            return True
        else:
            print(f"This size doesn't suits me")
            return False
    
    def like_brand_color(self, coat):
        if self.brand == coat.brand and self.color == coat.color:
            print(f"I like color and brand of this coat")
            return True
        else:
            print(f"I don't like this coat")
            return False
```

```python
class Customer(Vizitor):
    #аттрибуты порождаемого экземпляра класса
    def __init__(self, sex, size, brand, color, money_type: str = 'cash', money: int=0):
        self.size = size
        self.sex = sex
        self.brand = brand
        self.color = color
        self.money_type = money_type
        self.money = money

    #методы порождаемого экземпляра класса
    def buy(self, coat):
        chance = self.like_brand_color(coat) and self.put_on_coat(coat)
        if chance and self.money >= coat.price:
            print(f"I'll take it, here is the {self.money_type}")
            return True
        else:
            print(f"I can't take this coat. Bye!")
            return False
```
